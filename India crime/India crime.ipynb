<h1> ANALYSIS OF CRIME IN INDIA

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd
data = pd.read_csv('20_Victims_of_rape.csv')
data.head()
```

### Rape_Cases_Reported	and Victims_of_Rape_Total are the same so we remove the redundant column

```python
data = data.drop(['Victims_of_Rape_Total'],axis=1)
```

```python
data['Area_Name'].unique()
```

##  Number of rapes with respect to each state in 2001

```python
df = data[(data['Year'] == 2001) & (data['Subgroup'] == 'Total Rape Victims')]
plt.figure(figsize=(10,6))
sns.barplot(df['Area_Name'], df['Rape_Cases_Reported'], alpha=0.8)
plt.xticks(rotation='vertical')
plt.xlabel('States', fontsize=14)
plt.ylabel('Number os rapes', fontsize=14)
plt.title("Number of rapes with respect to each state", fontsize=16)
plt.show()
```

### Madhya pradesh leads in the number of rape cases followed by uttar pradesh and maharashtra

```python
max_rape = df[(df['Rape_Cases_Reported'] == max(df['Rape_Cases_Reported']))]
print("The most unsafe state for women in India is:",max_rape['Area_Name'].values[0])
```

```python
min_rape = df[(df['Rape_Cases_Reported'] == min(df['Rape_Cases_Reported']))]
print("The safest place for women in India is:",min_rape['Area_Name'].values[0])
```

```python
data['Year'].unique()
```

## Number of rapes in India from 2001 to 2010

```python
a=[]
for i in range(2001,2011):
    yr = data[(data['Year'] == i) & (data['Subgroup'] == 'Total Rape Victims')]
    a.append(sum(yr['Rape_Cases_Reported']))
plt.figure(figsize=(10,6))
plt.plot(data['Year'].unique(),a, color = 'red')
plt.title('Number of rapes in India from 2001 to 2010')
plt.xlabel('Year')
plt.ylabel('Number of rapes')
plt.show()   
```
